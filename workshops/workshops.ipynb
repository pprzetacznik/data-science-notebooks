````
def check_sum_on_dict_enumerate(my_list, n):
    bow = initialize_bow(my_list)
    for item in my_list:
        if (n - item) in bow:
            if (n - item) == item:
                if bow[n - item] > 1:
                    return item, item
                else:
                    return False
            else:
                return item, n - item
    return False
````

````
from math import sqrt

operations = {
    "*": (lambda a, b: a * b, 2),
    "+": (lambda a, b: a + b, 2),
    "-": (lambda a, b: a - b, 2),
    "/": (lambda a, b: a / b, 2),
    "pow2": (lambda a: a * a, 1),
    "sqrt2": (lambda a: sqrt(a), 1),
}


def calculate(my_list):
    stack = []
    for item in my_list:
        if item in operations:
            operation, params = operations[item]
            args = [stack.pop() for _ in range(params)]
            stack.append(operation(*reversed(args)))
        else:
            stack.append(item)
    return stack


def test_rpn():
    assert calculate([12, 2, 3, 4, "*", 10, 5, "/", "+", "*", "+"]) == [40]
    assert calculate([5, 1, 2, "+", 4, "*", "+", 3, "-"]) == [14]
    assert calculate([4, 1, "*", "pow2", "sqrt2"]) == [4]
````

````
class RPN:
    def __init__(self, operations=None):
        if not operations:
            self.operations = {
                "*": lambda a, b: a * b,
                "+": lambda a, b: a + b,
                "-": lambda a, b: a - b,
                "/": lambda a, b: a / b,
                "pow2": lambda a: a * a
            }
        else:
            self.operations = operations

    def calculate(self, my_list):
        stack = self._initialize_stack()
        for item in my_list:
            if item in self.operations:
                operation = self.operations[item]
                a, b = self._pop_2_from_stack(stack)
                result = operation(a, b)
                stack.append(result)
            else:
                stack.append(item)
        return stack

    def _pop_2_from_stack(self, stack):
        b = stack.pop()
        a = stack.pop()
        return a, b

    def _initialize_stack(self):
        return []


def test_rpn_simple():
    operations = {
        "*": lambda a, b: a * b,
        "+": lambda a, b: a + b,
        "-": lambda a, b: a - b,
        "/": lambda a, b: a / b
    }
    rpn = RPN(operations)
    assert rpn.calculate([2, 3, "+", 5, "*"]) == [25]


def test_rpn_complex():
    rpn = RPN()
    assert rpn.calculate([12, 2, 3, 4, "*", 10, 5, "/", "+", "*", "+"]) == [40]
    assert rpn.calculate([5, 1, 2, "+", 4, "*", "+", 3, "-"]) == [14]
````

````
class MyDuck:
    counter = 0

    def __init__(self, name):
        print(f"Duck {name} was born")
        self.name = name

    def fly(self):
        print(f"Duck {self.name} is flying")

    def kwa(self):
        print("Kwa kwa")

    @staticmethod
    def get_kind():
        return "Duck"

    @staticmethod
    def get_new_duck(name):
        if MyDuck.counter >= 1:
            return "No more ducks"
        else:
            MyDuck.counter += 1
            return MyDuck(name)

    @classmethod
    def bar(cls, b):
        print(f"{cls}, {b}")

    def bar2(self, b):
        print(f"{self.__class__}, {b}")

    def __eq__(self, other):
        return self.name == other.name

    def __call__(self):
        self.kwa()

    def __str__(self):
        return self.name

    def __repr__(self):
        return self.name

    def __add__(self, other):
        return MyDuck(f"{self.name}{other.name}")


my_duck = MyDuck("Donald")
if hasattr(my_duck, "kwa"):
    my_duck.kwa()
if hasattr(my_duck, "kwa2"):
    my_duck.kwa2()

my_ducks = [my_duck, 5, "asdfsdf"]
for duck in my_ducks:
    if hasattr(duck, "kwa"):
        duck.kwa()
````

````
# a game is over when all fields in a row are taken by a player

class TicTacToe:

    def __init__(self):
        self.board = self._initialize_board()

    def _initialize_board(self):
        return [['']*3 for i in range(3)]

    def print_board(self):
        for row in self.board:
            for i in row:
                print(f'|{i}', end='')
            print('|')

    # def is_game_over(self):
    #     return True

    def mark_x(self, x, y):
        self.board[x][y] = "x"

    def mark_o(self, x, y):
        self.board[x][y] = "o"

    def is_game_over(self):
        for player in self.players:
            for i in range(len(self.board)):
                is_over = True
                for j in range(len(self.board[0])):
                    if self.board[i][j] != player:
                        is_over = False
                if is_over:
                    return is_over
            for i in range(len(self.board[0])):
                is_over = True
                for j in range(len(self.board[0])):
                    if self.board[i][j] != player:
                        is_over = False
                if is_over:
                    return is_over
            is_over = True
            for i in range(len(self.board)):
                if self.board[i][i] != player:
                    is_over = False
                if is_over:
                    return is_over
            is_over = True
            for i in range(len(self.board)):
                if self.board[len(self.board) - i - 1][i] != player:
                    is_over = False
                if is_over:
                    return is_over
        return False


def test_raw_taken_by_player():
    tictactoe = TicTacToe()
    tictactoe.mark_x(0, 0)
    tictactoe.mark_x(0, 1)
    tictactoe.mark_x(0, 2)
    tictactoe.print_board()
    assert tictactoe.is_game_over() == True

def test_raw_taken_by_player_false():
    tictactoe = TicTacToe()
    tictactoe.mark_x(0, 1)
    tictactoe.mark_x(0, 2)
    assert tictactoe.is_game_over() == False
````

````
from functools import wraps


class depends:
    _run = {}

    def __init__(self, *deps):
        self.deps = deps

    def __call__(self, func):
        @wraps(func)
        def helper():
            if func not in depends._run:
                depends._run[func] = False
            for dep in self.deps:
                dep()
            if not depends._run[func]:
                func()
                depends._run[func] = True

        return helper


@depends()
def h():
    print("h()")


@depends(h)
def g():
    print("g()")


@depends(g, h)
def f():
    print("f()")


print(depends._run)
h()
print(depends._run)
f()
print(depends._run)
f()
f()
f()
print(depends._run)
````

````
import logging
from datetime import datetime
import os
import time
from functools import wraps


class Logger:
    __register = False

    def __init__(self):
        if not self.__register:
            self._init_default_register()

    def _init_default_register(self):
        if not os.path.isdir("logs"):
            os.makedirs("logs")
        logging.basicConfig(
            level=logging.DEBUG,
            format="%(asctime)s %(name)-12s %(levelname)-8s %(message)s",
            datefmt="%Y-%m-%d %H:%M:%S",
            filename=datetime.now().strftime("logs/logs_%Y_%m_%d_%H_%M.log"),
            filemode="w",
        )
        console = logging.StreamHandler()
        console.setLevel(logging.INFO)
        formatter = logging.Formatter(
            "%(name)-12s: %(levelname)-8s %(message)s"
        )
        console.setFormatter(formatter)
        logging.getLogger("").addHandler(console)
        logging.info("Logger initialized")
        Logger.__register = True

    def get_logger(self, filename):
        return logging.getLogger(filename)


def log(cls):
    cls.logger = Logger().get_logger(cls.__name__)
    return cls


def log_timer(fn):
    @wraps(fn)
    def helper(self, *args, **kw):
        start = time.time()
        result = fn(self, *args, **kw)
        self.logger.info(f"Invoking method {fn.__name__}() took {time.time() - start}s")
        return result
    return helper

@log
class Math:
    @log_timer
    def linear_function(self, x, a, b):
        self.logger.info(f"Method invoked with arguments: x = {x}, a = {a}, b = {b}")
        result = a * x + b
        return result

Math().linear_function(1, 1, 1)
````

### Composite pattern

https://pl.wikipedia.org/wiki/Kompozyt_(wzorzec_projektowy)

````
class Hardware:
    price = 0.0
    name = ""

    def __init__(self, name):
        self.name = name

    def __iter__(self):
        return iter([])

    def get_name(self):
        return self.name

    def get_price(self):
        return self.price


class CompositeHardware(Hardware):
    def __init__(self, name):
        super().__init__(name)
        self.items = []

    def get_price(self):
        return self.price + sum(item.get_price() for item in self.items)

    def add(self, item):
        self.items.append(item)

    def remove(self, item):
        self.items.remove(item)

    def __iter__(self):
        return iter(self.items)

    def get_name(self):
        result = [super().get_name()]
        for item in self.items:
            name = item.get_name()
            if isinstance(name, list):
                result += name
            else:
                result += [name]
        return result


class CPU(Hardware):
    price = 100.0


class GPU(Hardware):
    price = 140.0


class Mainboard(CompositeHardware):
    price = 120.0


class Casing(CompositeHardware):
    price = 50.0


if __name__ == "__main__":
    computer = Casing("Coolermaster Elite RC-331")
    mobo = Mainboard("Gigabyte GA-965P DS3")
    cpu = CPU("Intel E2160")
    gpu = GPU("Galaxy GeForce 7600GT 256MB DDR3 1,3ns 600/1500 Zalman")

    mobo.add(cpu)
    mobo.add(gpu)

    computer.add(mobo)

    print(f"Hardware: {computer.get_name()}")
    print(f"Price: {computer.get_price()}")
````

### Complex lambdas

````
def nwd(a, b):
     if b == 0:
         return a
     else:
         if b > a:
             return nwd(a, b % a)
         return nwd(b, a % b)

print(nwd(13,7))
print(nwd(9, 3))
print(nwd(26, 13))
````

````
def nwd(a, b):
     if b == 0:
         return a
     return nwd(b, a % b)

print(nwd(13,7))
print(nwd(9, 3))
print(nwd(26, 13))
````

````
def nwd(a, b):
    return a if b == 0 else nwd(b, a % b)

print(nwd(13,7))
print(nwd(9, 3))
print(nwd(26, 13))
````

### ABC classes and Data classes

* https://docs.python.org/3/library/dataclasses.html
* https://docs.python.org/3/library/abc.html
* https://stackoverflow.com/questions/3570796/why-use-abstract-base-classes-in-python/3571946

````
from dataclasses import dataclass, field

@dataclass(frozen=True)
class InventoryItem:
    """Class for keeping track of an item in inventory."""
    name: str
    unit_price: float
    quantity_on_hand: int = 0
    description: str = field(default="aaa", repr=False)

    def total_cost(self) -> float:
        return self.unit_price * self.quantity_on_hand

inventory_item = InventoryItem(name="book", unit_price="20")
inventory_item
````

````
from dataclasses import FrozenInstanceError

try:
    inventory_item.name = "milk"
except FrozenInstanceError as e:
    print(e)
````

````
from abc import ABCMeta, abstractmethod

class Duck(metaclass=ABCMeta):
    __slots__ = ()

    @abstractmethod
    def kwa(self, message):
        return False

    @classmethod
    def __subclasshook__(cls, C):
        if cls is Duck:
            if any("kwa" in B.__dict__ for B in C.__mro__):
                return True
        return NotImplemented

class DonaldDuck:
    def kwa(self, message):
        print(f"Kwa {message}")

class Dog:
    def wow(self):
        print("Wow")

print(issubclass(DonaldDuck, Duck))
print(isinstance(DonaldDuck(), Duck))
DonaldDuck().kwa("kwa")
````

````
print(issubclass(Dog, Duck))
print(isinstance(Dog(), Duck))
Dog().wow()
````
